```python
import pandas as pd
import numpy as np
import pyarrow.parquet as pq
```

```python
df = pd.read_csv('/Users/kdugg/Documents/zillow-short.csv')
```

```python
df = df.rename(columns = {'Unnamed: 0':'ZIP_CODE_SHORT'})
```

```python
df['ZIP_CODE_SHORT'] = df['ZIP_CODE_SHORT'].apply(lambda x: str(x).zfill(3))
df
```

```python
df = df[df['1/31/1996'].notna()]
```

```python
df
```

```python
df.groupby(['ZIP_CODE_SHORT']).mean()
```

```python
daycols = (pd.DatetimeIndex(df.columns[1:]) + pd.DateOffset(1)).tolist()
```

```python
daycol = ["{}/{}/{}".format(str(x.month).zfill(2), "01", x.year) for x in daycols]
```

```python
daycol.insert(0, "ZIP_CODE_SHORT")
```

```python
df.columns = daycol
```
