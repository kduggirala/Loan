```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/kdugg/Downloads/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
```

```python
months = ['-01-31', '-02-28', '-03-31', '-04-30', '-05-31', '-06-30', '-07-31', '-08-31', '-09-30', '-10-31', '-11-30', '-12-31']
for i in range(1996, 2000):
    for month in months:
        if i == 1996 and month == '-02-28':
            month = '-02-29'
        del df['{}{}'.format(i, month)]
```

```python
months = ['-01-31', '-02-28', '-04-30', '-05-31', '-07-31', '-08-31', '-10-31', '-11-30']
for i in range(2000, 2020):
    for month in months:
        if month == '-02-28' and i % 4 == 0:
            month = '-02-29'
        del df['{}{}'.format(i, month)]
```
