```python
def fix(l):
    include = [0]
    for i in range(1, len(l) - 1):
        if (l[i] - l[i-1])*(l[i+1]-l[i]) < 0:
            include.append(i)
    include.append(len(l) - 1)
    return [l[x] for x in include]

t = int(input())
for i in range(t):
    input()
    l = [int(x) for x in input().strip().split(' ')]
    f = fix(l)
    print(len(f))
    print(' '.join([str(x) for x in f]))
```
